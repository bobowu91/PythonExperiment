## Maximum Likelihood Estimation for Binomial Distribution ##
Only L(p) is needed for estimation  

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True,style='darkgrid')
```

```python
n = 20 # number of experiments/trials
p = 0.5 # success rate
k = np.random.binomial(n, p, 2000)
sns.distplot(k, kde = False)
```

```python
avg = np.mean(k)
std = np.std(k)
print(avg, std)
```

```python
def binomial(n,p,k):
    return (math.factorial(n)/(math.factorial(n-k) * math.factorial(k))) * (p ** k) * ((1-p) ** (n-k))
```

```python
def find_max(x, y): # input x,y are both lists; x is samples range and y is the log likelihood product
    maxi = -100000 # make sure this is not bigger than all values 
    max_index = 0
    for i, num in enumerate(y):
        if num > maxi:
            maxi = num
            max_index = i
    return maxi, x[max_index]
```

```python
ll = [ ] # likelihood list from p = 0 to 1
xaxis = [ ]
for i in range(5,99,5):
    p = i/100
    likelihood = [binomial(n, p, val) for val in k]
    log_transform = [math.log10(l) for l in likelihood]
    product = sum(log_transform)
    ll.append(product)
    xaxis.append(p)
```

```python
sns.scatterplot(x = xaxis, y = ll)
```

```python
loglikelihood, pmax = find_max(xaxis, ll)
print(pmax)
```

```python
ll = [ ] # likelihood list from n = 0 to 10
xaxis = [ ]
for i in range(n,n+20):
    p = pmax # from previous estimate
    likelihood = [binomial(i, p, val) for val in k]
    log_transform = [math.log10(l) for l in likelihood]
    product = sum(log_transform)
    ll.append(product)
    xaxis.append(i)
```

```python
sns.scatterplot(x = xaxis, y = ll)
```
